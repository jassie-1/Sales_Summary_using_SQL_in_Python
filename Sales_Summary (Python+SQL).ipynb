```python
import pandas as pd
import mysql.connector as sql
import matplotlib.pyplot as plt
```

```python
#Creating a Connection
conn= sql.connect(host= "localhost", user= "root", password= "", database= 'wow') #Private info, cannot tell
cur= conn.cursor()
```

```python
#Query to fetch the Product name, Total Quantity and Total Revenue
query= "SELECT Product_Name AS Product, SUM(Units_Sold) AS Quantity, SUM(Total_Revenue) AS Revenue FROM sales_data GROUP BY Product"
#Reading the query through pandas and printing it.
df= pd.read_sql_query(query, conn)
print(df)
```

```python
#Plotting bar chart 
df.plot(kind= 'bar', x= 'Product', y= 'Revenue', figsize= (50, 5))
#Rotating the name displayed in x-axis to 45 and ensuring the names are included in the saved image.
plt.xticks(rotation= 45, ha= 'right')
plt.tight_layout()
plt.savefig("C:\\Users\\DELL\\Desktop\\Internship\\Day-7\\Sales_Chart.png", bbox_inches= 'tight')
```
